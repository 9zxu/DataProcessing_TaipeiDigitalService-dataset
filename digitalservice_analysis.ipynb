##### drop NULL

```python
data=pd.DataFrame(df, columns=["編號","資料名稱","資料連結網址","資料集類型","最後更新時間","建立日期"])
data
```

##### ["資料集類型"] 參數

```python
data["資料集類型"].value_counts()
```

##### to data time

```python
data["最後更新時間"]=pd.to_datetime(data["最後更新時間"])
data["建立日期"]=pd.to_datetime(data["建立日期"])
data.info()
```
